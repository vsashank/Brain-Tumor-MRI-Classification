```python
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
train_path = "../dataset/Brain tumor/Training"
test_path = "../dataset/Brain tumor/Testing"

print("Training Folder :", train_path)
print("Testing Folder  :", test_path)
```

```python
classes = os.listdir(train_path)

print("Classes Found:")
print(classes)
```

```python
class_counts = {}

for cls in classes:

    folder = os.path.join(train_path, cls)

    class_counts[cls] = len(os.listdir(folder))

print(class_counts)
```

```python
for cls, count in class_counts.items():
    print(f"{cls:12} : {count} images")
```

```python
plt.figure(figsize=(12, 8))

for i, cls in enumerate(classes):

    folder = os.path.join(train_path, cls)

    image_name = os.listdir(folder)[0]

    image_path = os.path.join(folder, image_name)

    image = Image.open(image_path)

    plt.subplot(2, 2, i + 1)

    plt.imshow(image)

    plt.title(cls)

    plt.axis("off")

plt.tight_layout()

plt.show()
```

```python
for cls in classes:

    folder = os.path.join(train_path, cls)

    image_name = os.listdir(folder)[0]

    image_path = os.path.join(folder, image_name)

    image = Image.open(image_path)

    print(f"{cls:12} : {image.size}")
```

```python
train_total = 0

for cls in classes:
    train_total += len(os.listdir(os.path.join(train_path, cls)))

print("Total Training Images :", train_total)
```

```python
test_classes = os.listdir(test_path)

test_total = 0

for cls in test_classes:
    test_total += len(os.listdir(os.path.join(test_path, cls)))

print("Total Testing Images :", test_total)
```
